```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('data.csv')
data.dropna(inplace=True)
mean_quality = data.quality.mean()
is_good = (data.quality >= mean_quality).to_numpy()
data.drop(['Id','quality'],axis=1,inplace=True)
columns_names = list(data.columns)
```

```python
np_data = data.to_numpy()
np_data = StandardScaler().fit_transform(np_data)
pca = PCA(n_components=5)
pca_data = pca.fit_transform(np_data)
```

```python
# colors = ['blue' if good else 'red' for good in is_good]
# for i in range(5):
#     for j in range(5):
#         if (i != j):
colors = ['blue' if good else 'red' for good in is_good]
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=colors)
plt.xlabel(f'Componente Principal 1')
plt.ylabel(f'Componente Principal 2')
plt.title(
    'Análise de PCA - Dados no Espaço dos Componentes Principais')
plt.grid(True)
plt.show()
```
